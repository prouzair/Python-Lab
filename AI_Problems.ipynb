```python
import heapq

def uniform_cost_search(graph, start, goal):
    # Initialize the priority queue with the start node and its cost
    queue = [(0, start)]
    
    # Initialize the dictionary to store the cumulative costs
    cumulative_costs = {start: 0}
    
    # Initialize the dictionary to store the parent nodes
    parents = {start: None}
    
    while queue:
        # Pop the node with the lowest cost from the priority queue
        cost, current_node = heapq.heappop(queue)
        
        if current_node == goal:
            # Reached the goal node, construct the path and return it
            return construct_path(parents, current_node)
        
        # Explore the neighbors of the current node
        for neighbor, neighbor_cost in graph[current_node].items():
            new_cost = cumulative_costs[current_node] + neighbor_cost
            
            if neighbor not in cumulative_costs or new_cost < cumulative_costs[neighbor]:
                # Update the cumulative cost and parent node for the neighbor
                cumulative_costs[neighbor] = new_cost
                parents[neighbor] = current_node
                
                # Add the neighbor to the priority queue with the updated cost
                print(new_cost,neighbor)
                heapq.heappush(queue, (new_cost, neighbor))
              
    
    # No path found
    return None

def construct_path(parents, goal):
    # Reconstruct the path from the goal to the start node
    path = [goal]
    while parents[goal]:
        goal = parents[goal]
        path.append(goal)
    
    # Reverse the path to get the correct order
    return list(reversed(path))



graph = {
    
          'S':{'A':5,'B':9,'D':6},
          'A':{'B':3,'G1':9},
          'B':{'A':3,'C':1},
          'C':{'F':7,'G2':5,'S':6},
          'D':{'C':2,'E':2,'S':1,},
          'E':{'G3':7,},
          'F':{'D':2,'G3':8},
          'G1':{},
          'G2':{},
          'G3':{}    
         }

solution  = uniform_cost_search(graph,'S','G2')
print(solution)
```
